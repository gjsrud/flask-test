```python
import seaborn as sns
import pandas as pd
```

```python
import sklearn
```

```python
#타이타닉 불러오기
titanic=sns.load_dataset('titanic')

#일단 df라는 분석 원본 만들기
df = titanic.loc[:, ['age','sex','fare','class']]
df
```

```python
#문자값을 숫자값으로 치환
df_data=df.replace({'male': 1, 'female': 2, "Third":3, "Second" :2, "First":1})
df_data
```

```python
#결측값 있는지 확인
df_data.isna().sum()
```

```python
#평균구할라고 import
import pandas as mean
```

```python
#결측치를 평균으로 치환해서 해보자!
#평균부터 구하고
df_data.mean()
```

```python
# 중위수!
df_data.median()
```

```python
#결측치는 중위값으로 맞춰주고
df_data=df_data.fillna(28.0)
```

```python
df_data
```

```python
#결측치 있나? 없죠?!
df_data.isna().sum()
```

```python
#살았는지만 따로 떼서보자 -> 이게 붓꽃데이터에서의 꽃이름이라고 보면 된다
df_sur = titanic.loc[:,['survived']]
df_sur
```

```python
from sklearn.tree import DecisionTreeClassifier #분류용!
from sklearn.model_selection import train_test_split #train, test분리
```

```python
#원본 바뀌는 거 방지
df_test=df_data.copy()
```

```python
#원본이랑 survived더해라!
df_test['survived'] = df_sur
df_test
```

```python
#중요! 테스트랑 훈련 데이터 분리하는 거!!
#random_state=11 이게 정확도가 제일 높네
X_train,X_test,y_train,y_test = train_test_split(df_data, df_sur,test_size=0.25,random_state=11)
```

```python
dt_clf = DecisionTreeClassifier(random_state=11)
dt_clf.fit(X_train,y_train)
pred = dt_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

```python
#나이, 성별, 요금, 클래스
dt_clf.predict([[32.0,2,51.000,2]])
```
